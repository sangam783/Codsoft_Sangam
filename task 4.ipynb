SPLITTING FOR TESTING

```python
X = df['TV']
y = df['Sales']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.7, test_size = 0.3, random_state = 100)
```

BUILDING LINEAR MODEL

```python
X_train.head()
```

```python
y_train.head()
```

```python
import statsmodels.api as sm
```

```python
X_train_sm = sm.add_constant(X_train)
lr = sm.OLS(y_train, X_train_sm).fit()
```

```python
lr.params
```

```python
print(lr.summary())
```

DATA VISUALISATION

```python
advertising = pd.read_csv('advertising.csv')

fig, ax = plt.subplots(figsize=(12, 6))

ax.scatter(df['TV'], df['Sales'], label='TV')
ax.scatter(df['Newspaper'],df['Sales'], label='Newspaper')
ax.scatter(df['Radio'], df['Sales'], label='Radio')

ax.set_xlabel('Advertising')
ax.set_ylabel('Sales')

ax.legend()

plt.show()
```

```python
corr_matrix = df.corr()

fig, ax = plt.subplots(figsize=(8, 6))
heatmap = ax.imshow(corr_matrix, cmap='YlGnBu')

for i in range(len(df.columns)):
    for j in range(len(df.columns)):
        ax.text(j, i, f"{corr_matrix.iloc[i, j]:.2f}",
                ha='center', va='center', color='black')


plt.title('Correlation Heatmap')
plt.show()
```
